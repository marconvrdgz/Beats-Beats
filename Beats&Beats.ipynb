* Se sabe que las pulsaciones por minuto del ser humano comienzan en un valor alto cuando se es joven y disminuyen a medida que el ritmo cardíaco se ralentiza, ya que el cuerpo experimenta un proceso natural de envejecimiento y autodegradación.

* Los BPM (Beats Per Minute o Pulsos Por Minuto) es una unidad de medida utilizada para cuantificar el ritmo musical y la velocidad del ritmo en una pieza musical de diversos géneros. Están estrechamente relacionados con el tempo de la canción. De hecho, el tempo de una canción se expresa comúnmente en términos de BPM.

          |    Tempo       |    BPM Range    |

          | Prestissimo    | Más de 200      |
          | Presto         | 168-200         |
          | Vivace         | 132-168         |
          | Allegro        | 120-168         |
          | Allegretto     | 112-132         |
          | Moderato       | 108-120         |
          | Andante        | 76-108          |
          | Adagietto      | 66-76           |
          | Adagio         | 56-76           |
          | Lento          | 45-56           |
          | Largo          | Menos de 45     |

##### Continuamos con la revision y limpieza de los datos. Deacuerdo con la informacion del sitio del DataSet "Spotify_Youtube" las categorias que nos interesan se describen como a continuacion.



*   Danceability: Describe qué tan adecuado es una pista para bailar basado en una combinación de elementos musicales como el tempo, la estabilidad del ritmo, la fuerza del ritmo y la regularidad general. Un valor de 0.0 es el menos bailable y 1.0 es el más bailable.

*   Energy: Es una medida de 0.0 a 1.0 y representa una medida perceptual de intensidad y actividad. Las pistas energéticas suelen sentirse rápidas, fuertes y ruidosas. Por ejemplo, el death metal tiene alta energía, mientras que un preludio de Bach obtiene una puntuación baja en la escala. Las características perceptuales que contribuyen a este atributo incluyen el rango dinámico, la sonoridad percibida, el timbre, la tasa de inicio y la entropía general.

*   Tempo: El tempo estimado general de una pista en beats por minuto (BPM). En términos musicales, el tempo es la velocidad o ritmo de una pieza dada y se deriva directamente de la duración promedio del compás.

#####Los atributos del DataSet heart incluyen los siguientes:

*   Edad: edad del paciente [años]

*   Tipo de dolor en el pecho: tipo de dolor en el pecho [TA: Angina Típica, ATA: Angina Atípica, NAP: Dolor no Anginal, ASY: Asintomático]

*   Angina inducida por ejercicio: angina inducida por ejercicio [Y: Sí, N: No]
*   Depresión antigua: depresión antigua = ST [Valor numérico medido en depresión]
*   Pendiente de ST: la pendiente del segmento ST en el ejercicio máximo [Up: ascendente, Flat: plana, Down: descendente]
*   Enfermedad cardíaca: clase de salida [1: enfermedad cardíaca, 0: Normal]
